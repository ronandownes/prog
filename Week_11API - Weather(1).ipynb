# Alternate API

https://open-meteo.com/en/docs/historical-weather-api

```python
endpoint = "https://archive-api.open-meteo.com/v1/era5?"
long = lon
lat = lat
start_date = "2022-10-25"
end_date= "2022-11-23"
params_list = [
    "temperature_2m",
    "rain",
    "direct_normal_irradiance",
    "soil_temperature_0_to_7cm",
    "soil_moisture_0_to_7cm"
]
params = ""
for param in params_list:
    params += param + ","
params = params[:-1]
query = f"latitude={lat}&longitude={long}&start_date={start_date}&end_date={end_date}&hourly={params}"

url = endpoint + query

req = requests.get(url)
```

```python
res = req.json()
```

```python
met_url = "https://metwdb-openaccess.ichec.ie/metno-wdb2ts/locationforecast"
lat = lat
long = lon
From = "2022-11-29T07:00"
to = "2022-11-29T17:00"
met_api = met_url + f"?lat={lat};long={long};to={to};from={From}"
req = requests.get(met_api)
```
