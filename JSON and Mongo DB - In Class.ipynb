```python
# Query, fields (columns)
res = coll.find({}, {})
for i in res:
    print(i)
```

```python
# Query, fields (columns)
res = coll.find({"name": "Sean"}, {})
for i in res:
    print(i)
```

```python
# Query, fields (columns)
res = coll.find({"Student Number": {"$gt": 15}}, {})
for i in res:
    print(i)
```

```python
# Query, fields (columns)
res = coll.find({"Student Number": {"$gt": 15}}, {'name':1, "_id":0})
for i in res:
    print(i)
```

```python
# Query, fields (columns)
res = coll.find({"name": {"$gt": "M", "$lt": "S"}}, {'name':1, "_id":0})
for i in res:
    print(i)
```

```python
# Query, fields (columns)
res = coll.find({"name": {"$eq": "Imran", "$lt": "S"}}, {'name':1, "_id":0})
for i in res:
    print(i)
```

```python
# Query, fields (columns)
res = coll.find({"name": {"$regex": "[Ss]"}}, {'name':1, "_id":0})
for i in res:
    print(i)
```
