```python
endpoint ="https://api.open-meteo.com/v1/forecast?"
lat = 53.33
long = -6.25
past_days = 7

params = [
    "temperature_2m",
    "rain",
    "soil_temperature_0cm",
    "soil_moisture_0_1cm",
    "direct_normal_irradiance",
]

values = ""
for param in params:
    values += param + ","
values = values[:-1]

url = f"{endpoint}latitude={lat}&longitude={long}&hourly={values}&past_days={past_days}"
```

```python
url
```

```python
import requests
```

```python
req = requests.get(url)
```

```python
req
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(req.json())
```
