# Merge

```python
import pandas as pd
```

```python
climate_temp = pd.read_csv("climate_temp.csv")
climate_precip = pd.read_csv("climate_precip.csv")
```

```python
climate_temp.head()
```

```python
precipStation1 = climate_precip[climate_precip["STATION"] ==  "GHCND:USC00049099"]
```

```python
innerMerge = pd.merge(precipStation1, climate_temp)
```

```python
innerMerge.head()
```

```python
print(climate_temp.shape)
print(precipStation1.shape)
print(innerMerge.shape)
```

```python
outer_merge = pd.merge(precipStation1, climate_temp, how="outer", on=["STATION", "DATE"])
```
