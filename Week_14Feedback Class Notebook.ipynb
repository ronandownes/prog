```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
em = pd.read_csv("Exxon Mobil.csv")
```

```python
em.head()
```

```python
fig, ax = plt.subplots()

points = ax.scatter("Revenue (in millions)", "Profit (in millions)", data=em, c="Year")
plt.xlabel("Revenue (in millions)")
plt.ylabel("Profit (in millions)")
plt.title("Revenue vs Profit for Exxon Mobil")
fig.colorbar(points)
plt.show()
```

```python
gm = pd.read_csv("General Motors.csv")
```

```python
gm.head()
```

```python
def companyScatter(df, colour='magma'):
    # Doc string
    
    fig, ax = plt.subplots()
    
    points = ax.scatter(df.columns[2], df.columns[3], data=df, c="Year", cmap=colour)
    plt.xlabel("Revenue (in millions)")
    plt.ylabel("Profit (in millions)")
    plt.title(f"Revenue vs Profit for {df['Company'][0]}")
    fig.colorbar(points)
    plt.show()
```

```python
companyScatter(em)
```

```python
companyScatter(gm, "magma")
```

```python
companies = [em, gm]
```

```python
mot = pd.read_csv("Motorola.csv")
```

```python
companies.append(mot)
```

```python
companies
```

```python
for company in companies:
    companyScatter(company, "magma")
```
